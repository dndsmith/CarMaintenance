```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("Mazda3MPG.csv")
df = df.iloc[:,:6]
df
```

```python
# plt.plot(df.to_numpy()[:,0], df.to_numpy()[:,3])
plt.plot(df["Date"], df["MPG"])
```

```python
import matplotlib.ticker as ticker

tick_spacing = 282

fig, ax = plt.subplots(1,1)
ax.plot(pd.to_datetime(df["Date"]), df["MPG"], label='_nolegend_')
ax.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))

# ADD MAJOR WORK

# 1. Changed Tires
x = ["2017-08-01", "2017-08-01"]
x = pd.to_datetime(x)
y = [20, 44]
ax.plot(x,y)

# 2. Changed Spark Plugs
x = ["2018-06-01", "2018-06-01"]
x = pd.to_datetime(x)
y = [20, 44]
ax.plot(x,y)

ax.set_ylabel("Miles Per Gallon", size="large", weight="semibold", labelpad=15)
ax.set_xlabel("Recorded 2016-08-10 through 2020-03-15", size="large", weight="semibold", labelpad=15)
ax.legend(["Changed Tires", "Changed Spark Plugs"])
ax.set_title("Fuel Efficiency 2012 Mazda2 SkyActiv", size="x-large", weight="semibold", pad=10)
current_fig = plt.gcf()
```

```python
# save figure
current_fig.set_size_inches(12,8)
current_fig.savefig("visualization.png", bbox_inches='tight', dpi=100)
```
